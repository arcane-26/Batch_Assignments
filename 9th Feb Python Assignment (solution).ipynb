# QUESTION 3 (SOLUTION)

when the child class inherit the properties and methods of two of more parent class then this type of inhertance is known as multiple inheritance

```python
class father:
    def knowledge(self):
        return "got knowledge from father"
    
class mother:
    def beauty(self):
        return "got beauty from mother"
    
class child(father,mother):
    pass
```

```python
daughter=child()
```

```python
daughter.knowledge()
```

```python
daughter.beauty()
```

# QUESTION 4 (SOLUTION)

 Getters and setter methods are used to access and make changes to the private attributes of the class, when certain certain conditon are met which are present in these methods

```python
class create_pin:
    def __init__(self,pin):
        self.__pin=int(pin)
        
        
    def get_pin(self):            # getter method
        return self.__pin
    
    def change_pin(self,new_pin):   # setter method
        if type(new_pin)==int:
            self.__pin=new_pin
        else:
            return "you cannot change pin"
            
        
```

```python
atm=create_pin(2112)
```

```python
atm.get_pin()
```

```python
atm.change_pin("hehe")
```

```python
atm.change_pin("1232")
```

```python
atm.change_pin(1011)
```

```python
atm.get_pin()
```

# QUESTION 5 (SOLUTION)

wthe object of child class want to access the method with a name which is present in both child and parent class, hence in this case method of parent class is overrided and the object of child class will access its own method

```python
class phone:
    
    def display(self):
        return "Inside phone class"
    
class samsung(phone):
    
    def display(self):
        return "Inside samsung class"
```

```python
p=samsung()
```

```python
p.display()
```
