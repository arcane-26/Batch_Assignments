# QUESTION 2 (SOLUTION)

Some of the key features of MongoDB include:

Document-oriented: MongoDB is a document-oriented database, which means that data is stored in documents, rather than in tables with rows and columns. Each document is stored as a JSON-like object, which makes it easy to work with data that has a flexible schema.

Scalability: MongoDB is designed to be highly scalable and can handle large volumes of data with ease. It supports automatic sharding, which allows you to distribute data across multiple servers to achieve horizontal scaling.

Flexible data model: MongoDB's flexible data model allows you to store data of different types and structures in the same collection, making it easy to handle data that may not fit neatly into a rigid schema.

Indexing: MongoDB supports a range of indexing options, including single-field, compound, and geospatial indexes. This allows for fast and efficient querying of data.
