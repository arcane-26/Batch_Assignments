```python
# example of raise

class custom_exception(Exception):
    def __init__(self,msg):
        self.msg=msg
```

```python
def raise_error(m):
    if len(m)!=10:
        raise custom_exception("mobile number is not valid")
    else:
        raise custom_exception("valid number")
```

```python
try:
    n=input("Enter your mobile number: ")
    raise_error(n)
    print(n)
except custom_exception as e:
    logging.error(f"mobile no. should consist 10 digits --->{e}")
```

```python
class custom_exception(Exception):
    def __init__(self,msg):
        self.msg=msg
```

```python
def parse(m):
    if m[0].lower()!='m' or m[0].lower!='f' or m[0].lower!='O':
        raise custom_exception("Invalid Gender")
```

```python
try:
    gender=input(''' Enter gender: ''')
    parse(gender)
except custom_exception as e:
    logging.error(f'Your gender can be either Male,Female or Other---->{e}')
```
