# QUESTION 1 (SOLUTION)

The GET method is used to retrieve data from a server. When a GET request is sent to a server, the server responds with the requested data, which is typically displayed in the user's web browser. GET requests are often used for simple queries or searches, as they do not change any data on the server. In a GET request, the data is sent as part of the URL, in the form of query parameters, which can be seen in the browser's address bar.

The POST method is used to submit data to a server, typically for the purpose of changing or updating data. When a POST request is sent to a server, the data is sent in the request body, rather than as part of the URL. This means that the data is not visible in the browser's address bar. POST requests are often used for form submissions or other types of data updates.

With the request object, Flask allows developers to extract information such as query parameters, form data, and request headers from a request. This information can then be used by the application to generate an appropriate response to the client.

For example, a Flask web application may use the request object to:

Retrieve the contents of a form submitted by a user,
Access query parameters in a GET request,
Extract the user's IP address or browser information,
Retrieve cookies sent by the client

The redirect() function in Flask is used to redirect the user to a different endpoint or URL. It is a convenient way to redirect users to a different page or resource based on some condition or user action.

In Flask, a template is a file that contains a mixture of HTML, CSS, and placeholders for dynamic content that will be generated by the application. A template is used to generate the HTML that will be sent to the client, allowing for dynamic and customizable pages.

The render_template() function is used to render a template file and generate HTML content. It takes the name of the template file as an argument, and any additional variables that need to be passed to the template for rendering.
