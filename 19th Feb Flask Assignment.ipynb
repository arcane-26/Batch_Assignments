Flask is a popular open-source web framework written in Python. It is a lightweight and flexible framework that allows developers to quickly build web applications with minimal code. Flask follows the Model-View-Controller (MVC) architecture and provides various tools and libraries that simplify the development process.

Advantages of Flask Framework

Lightweight and flexible: Flask is a lightweight and flexible web framework that allows developers to build applications quickly and efficiently without being constrained by rigid structures.

Easy to learn and use: Flask has a simple and intuitive syntax that is easy to learn and use, making it a popular choice for beginners and experienced developers alike.

Modular and extensible: Flask has a modular architecture and supports a wide range of extensions that can be easily integrated with the framework, making it highly customizable and extensible.

Built-in development server: Flask comes with a built-in development server that makes it easy to test and debug applications during the development phase.

Compatible with various databases: Flask supports a variety of databases, including SQL-based databases like MySQL, PostgreSQL, and SQLite, as well as NoSQL databases like MongoDB.

# QUESTION 3 (SOLUTION)

In Flask, app routing is the process of mapping URLs to specific functions or views within a Flask application. When a user makes a request to a specific URL in the application, the Flask routing system matches the requested URL to a corresponding function or view, which generates an HTTP response that is returned to the user.

App routes in Flask are used to define the URLs that a Flask application can handle, and to specify the corresponding functions or views that should be executed when a user requests a particular URL. App routes allow developers to create a well-structured and organized Flask application by separating different parts of the application into different views or functions.
